Положим m=1, h=1 и k=1 для удобства, тогда зависимость энергии от n выглядит следующим образом:

```python
n=15
N = np.linspace(1, n-1, n)
plt.plot(N, E(1,1,n-1),'*', c='red',  label='E(n)')
plt.xlabel('n')
plt.ylabel('E')
plt.legend()
plt.suptitle('Энергетический спектр')
```

Видно, что растояние между соседними энергетическими уровнями уменьшается при увеличении n, что подтверждается аналитическим расчетом

```python
def Psi(k,m,n,x):
    b=(2*m*k)**(1/3)
    e=E(k,m,n)[n]
    arg=b*(x-e/k)
    Psi=sp.airy(arg)[0]
    return Psi
```

Завивисимость первых волновых функций от x выглядит так:

```python
x = np.linspace(0, 10, 1000)
for n in range(4):
    f=[]
    for i in range(1000):
        f.append(Psi(1,1,n,x[i]))
    plt.plot(x, f,  label=n)
    plt.xlabel('x')
    plt.ylabel('Psi')
    plt.legend()
    plt.suptitle('Волновые функции при различных n')
```

Видно, что функции как и ожидалось пересекают ось X ровно столько раз сколько и предсказывает осциляционная теорема (кол-во пересечений равно номеру функции). Так же заметно стремление к 0 ф-ции на бесконечности, что и предсказывает аналитическое решение.
